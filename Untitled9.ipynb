1. Объясните, как вычисляется качество модели с задачей классификации: назовите основные и вторичные метрики, приведите пример.

Ответ:

Качество модели определяется с помощью метрик.

Accuracy — доля правильных ответов от общего числа ответов. Правильные ответы — это истинно положительные (True Positive) и истинно отрицательные ответы (True Negative).

Precision — доля правильно классифицированных положительных образцов от всех образцов, которые были классифицированы как положительные.

Recall — Доля правильно классифицированных положительных образцов от общего числа положительных образцов класса.

F1-мера - взвешенное среднее гармоническое между precision и recall

Пример: В задаче классификации спама, если из 1000 сообщений 900 действительно являются спамом и 100 — не спамом, а модель правильно классифицировала 850 сообщений как спам и 90 сообщений как не спам, то:

- Accuracy: (90 + 850)/(90 + 50 + 850 + 10) = 0.94
- Precision: 90/(90 + 50) = 0.64
- Recall: 90/(90 + 10) = 0.9
- F1-мера: 2 * (0.64 * 0.9)/(0.64 + 0.9) = 0.75

2. Что такое тестовый набор данных (test set) и для чего он нужен?

Ответ:

Тестовый набор данных — это подмножество данных, которое используется для оценки производительности модели после её обучения. Он должен быть независимым от тренировочного набора данных, чтобы обеспечить объективную оценку качества модели на данных, которые она не видела в процессе обучения. Тестирование модели помогает определить, насколько хорошо она будет работать на новых данных, и помогает избежать переобучения.

3. В чем разница между параметром модели и гиперпараметром алгоритма обучения? Приведите примеры.

Ответ:

Параметры модели - это значения, которые модель учит во время обучения на основе данных. Они определяют поведение модели и её способность делать предсказания.

Пример: В линейной регрессии это коэффициенты (веса) для каждой из входных переменных.

Гиперпараметры алгоритма обучения - это параметры, которые устанавливаются до начала обучения модели и не изменяются в процессе обучения. Они влияют на то, как модель обучается, но сами по себе не обучаются.

Пример: Коэффициент обучения в алгоритмах оптимизации, таких как градиентный спуск или регуляризация (например, коэффициент L2-регуляризации).

Таким образом, параметры модели обучаются и оптимизируются на данных, в то время как гиперпараметры задаются до обучения и определяют, как будет происходить процесс обучения.

Разделим набор данных на 2 выборки в соотношении 60/40

```python
from sklearn.model_selection import train_test_split
features = train_data.drop(['Survived', 'Name', 'Ticket', 'Cabin'], axis=1).columns
X, y = train_data[features], train_data['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.4)
print('Train:', X_train.shape)
print('Test:', X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

model = RandomForestClassifier(n_estimators=27, max_depth=21, random_state=42)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

print(metrics.classification_report(y_train, y_train_pred))
print(metrics.classification_report(y_test, y_test_pred))
```

При нынешних параметрах RandomForestClassifier модель сильно переобучается на тренировочной выборке

Подберем оптимальные значения параметров для алгоритма

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [20, 28, 50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

grid_search.fit(X_train, y_train)

print(f"Лучшие параметры: {grid_search.best_params_}")
```

```python
model = RandomForestClassifier(min_samples_leaf= 4, n_estimators= 50, random_state=42)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

print(metrics.classification_report(y_train, y_train_pred))
print(metrics.classification_report(y_test, y_test_pred))
```

Получилось избавить модель от переобучения

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Предскажем на стандартизированных данных

Как и ожидалось, алгоритм не чувствителен к масштабированию, т.к. не взвешивает признаки

```python
from sklearn.decomposition import PCA

# создадим объект класса PCA
pca = PCA(n_components=6)
pca.fit(X_train_scaled)
# уменьшим размерность данных
train_x_pca = pca.transform(X_train_scaled)
test_x_pca = pca.transform(X_test_scaled)
```

Попробуем предсказать на уменьшенном количестве данных
